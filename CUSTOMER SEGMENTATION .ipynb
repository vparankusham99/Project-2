```python
#Removing the null values 

df = df.dropna()
print ("The total number of data after removing the missing values are:"),len(df)
```

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
dates = []

for i in df["Dt_Customer"]:
    i = i.date()
    dates.append(i)
```

```python
print("The newest customers enrollment date:",max(dates))
print("The oldest customers enrollment date:",min(dates))
```

```python
#Creating a new feature 

days = []
d1 = max(dates) #taking it to be the newest customer
for i in dates:
    delta = d1 - i
    days.append(delta)
df["Customer_For"] = days
df["Customer_For"] = pd.to_numeric(df["Customer_For"], errors="coerce")
```

```python
print("Total categories in the feature Marital_Status:\n", df["Marital_Status"].value_counts(), "\n")
print("Total categories in the feature Education:\n", df["Education"].value_counts())
```

# FEATURE ENGINEERING

```python
df["Age"] = 2022-df["Year_Birth"] #current age of customer.
```

```python
#creating a feature Spent = Indicates the amount spent by the customer
#in various categories over the span of two years.


df["Spent"] = df["MntWines"]+ df["MntFruits"]+ df["MntMeatProducts"]+ df["MntFishProducts"]+ df["MntSweetProducts"]+ df["MntGoldProds"]
```

```python
#Creating a feature called Living_With from the Marital_Status to extract the
#living situation of couples.

df["Living_With"]=df["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})
```

```python
#Creating a feature the indicates the total children living in that household

df["Children"]=df["Kidhome"]+ df["Teenhome"]
```

```python
# Creating a feature indicating the total number of members in the family.

df["Family_Size"] = df["Living_With"].replace({"Alone": 1, "Partner":2})+ df["Children"]
```

```python
#Creating a feature Is_Parent to indicate the parenthood status of the person.

df["Is_Parent"] = np.where(df.Children> 0, 1, 0)
```

```python
#Creating the categories according to the education level in a variable called Education.

df["Education"]=df["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})

```

```python
#Renaming some of the variables for easier usage.

df=df.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
```

```python
#Dropping some of the features which are not needed.

to_drop = ["Marital_Status", "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
df = df.drop(to_drop, axis=1)
```

# DATA PREPROCESSING

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import warnings
import sys
if not sys.warnoptions:
    warnings.simplefilter("ignore")
np.random.seed(42)
```

```python
#List of categorical columns.

s = (df.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables in the dataset:", object_cols)
```

```python
#Label Encoding the object datatypes.


LE=LabelEncoder()
for i in object_cols:
    df[i]=df[[i]].apply(LE.fit_transform)
```

```python
#Creating a copy of df.

df1 = df.copy()
```

```python
# creating a subset of dataframe by dropping the features on deals accepted
#and promotions


cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
df1 = df1.drop(cols_del, axis=1)
```

```python
#Scaling of data.

scaler = StandardScaler()
scaler.fit(df1)
scaled_df1 = pd.DataFrame(scaler.transform(df1),columns= df1.columns )


#The scaled data will be used for dimensionality reduction.
```

# Dimensionality Reduction


```python
#Dimensionality reduction is the process of reducing the number of random variables under consideration,
#by obtaining a set of principal variables.
```

```python
#Dimensionality Reduction with Principal Component Analysis
#PCA reduces the dimensionality of data, increasing interpretability but at the same time minimizing information loss.

#Reducing the dimension to 3.

pca = PCA(n_components=3)
pca.fit(scaled_df1)
PCA_df1 = pd.DataFrame(pca.transform(scaled_df1), columns=(["col1","col2", "col3"]))
PCA_df1.describe().T
```

```python
#3D Projection of data in the reduced dimension.

x =PCA_df1["col1"]
y =PCA_df1["col2"]
z =PCA_df1["col3"]


#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

# MODEL BUILDING

```python
#!pip install yellowbrick
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt, numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import AgglomerativeClustering
```

```python
#Using Elbow method to find the nuber of clusters to make

print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_df1)
Elbow_M.show()
```

```python
#It shows that forming four clusters would be optimum
```

```python
#Agglomerative Clustering Model

AC = AgglomerativeClustering(n_clusters=4)
```

```python
#Fitting the model

yhat_AC = AC.fit_predict(PCA_df1)
PCA_df1["Clusters"] = yhat_AC
```

```python
#Adding the Clusters column to the original dataset

df1["Clusters"]= yhat_AC
```

```python
#Plotting the clusters in 3D Distribution..

fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_df1["Clusters"], marker='o', cmap = cmap )
ax.set_title("The Plot Of The Clusters")
plt.show()
```
